```python
def identify_type(value):

    if ((isinstance(value, int)==True) or (isinstance(value, float)==True)):

        return "Number"
    
    elif (isinstance(value, str)==True):

        try:
            value=value.replace(",", ".")

            value=int(value)

            return "Number"
        
        except:
            return "String"

    elif (isinstance(value, bool)==True):

        return "Boolean"

int(123.66)
```
